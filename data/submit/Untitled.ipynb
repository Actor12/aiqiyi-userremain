```python
import pandas as pd

dnn = pd.read_csv('submit_B.csv',header=None)

dnn.head()
```

```python
dnn[dnn.iloc[:,1]>7.5]
```

```python
def deal(x):
    if x<0.5:
        return 0
    elif x>=0.5 and x<1.5:
        return 1
    elif x>=1.5 and x<2.5:
        return 2
    elif x>=2.5 and x<3.5:
        return 3
    elif x>=3.5 and x<4.5:
        return 4
    elif x>=4.5 and x<5.5:
        return 5
    elif x>=5.5 and x<6.47:
        return 6
    elif x>=6.47:
        return 7

res = dnn.apply(lambda x:deal(x[1]),axis=1)
res =pd.concat([dnn.iloc[:,0],res],axis=1)
res.head()
```
